Seperating x & y

```python
x = df[['Height']] # ind var
y = df['Weight']  # dep var
```

```python
x = df.drop('rent', axis = 1)
x = df[['Height']] # ind var
```

Spliting the data

```python
from sklearn.model_selection import train_test_split as tts
```

```python
df.head(10)
```

```python
xtrain, xtest, ytrain, ytest = tts (x,y,train_size=.70, random_state=1) 
```

```python
xtrain.shape
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()   # Object Creation
```

```python
lr.fit(xtrain ,ytrain)   # fitting the model
```

```python
xtrain.shape
```

Finding m & c value as per regression formula y= mx+c; m= coef, c= intercept

```python
lr.coef_
```

```python
lr.intercept_
```

```python
m = lr.coef_
c = lr.intercept_
```

Random Weight Predict as per formula

```python
x = 74.110105
y = (m * x) + c
y
```
